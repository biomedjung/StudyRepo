```python
def bmi_category(bmi):
  if bmi < 18.5:
    bmi_category = 'Underweight'
  elif bmi < 25:
    bmi_category = 'Normal Weight'
  elif bmi < 30:
   bmi_category = 'Overweight'
  elif bmi > 30:
    bmi_category = 'Obesity'
  return bmi_category

print(bmi_category(bmi))

```

# Max Hearth Rate (MHR)
