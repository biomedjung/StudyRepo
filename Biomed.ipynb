```python
nome = input('Qual seu nome?')
idade = input('Qual a sua idade?')
peso = input('Qual seu peso?')
print(nome, idade, peso)
```

```python
class Paciente:
    def __init__(self, nome, idade, genero, altura_cm, peso):
        self.nome = nome
        self.idade = idade
        self.genero = genero
        self.altura = altura_cm / 100
        self.peso = peso

    def calcular_imc(self):
        return self.peso / (self.altura ** 2)

pacientes = []

while True:
    nome = input("Digite o nome do paciente (ou 'fim' para sair): ")
    if nome == "fim":
        break

    idade = int(input("Digite a idade do paciente: "))
    genero = input("Digite o gênero do paciente (M/F): ")
    altura_cm = float(input("Digite a altura do paciente (em centímetros): "))
    peso = float(input("Digite o peso do paciente (em quilogramas): "))

    paciente = Paciente(nome, idade, genero, altura_cm, peso)
    pacientes.append(paciente)

    imc = paciente.calcular_imc()
    print(f"O IMC do paciente {paciente.nome} é {imc:.2f}.")


    def faixa_imc(self):
        imc = self.calcular_imc()
        if imc < 18.5:
            return "Abaixo do peso"
        elif 18.5 <= imc < 25:
            return "Peso normal"
        elif 25 <= imc < 30:
            return "Acima do peso"
        else:
            return "Obesidade"

print("Programa encerrado.")
```

```python
import psycopg2
import pandas as pd

# Conexão com o banco de dados
conn = psycopg2.connect(
    host="localhost",
    database="banco_de_dados",
    user="usuario",
    password="senha"
)

# Leitura dos dados da tabela
query = "SELECT * FROM hemograma"
df = pd.read_sql(query, conn)

# Cálculo de estatísticas
media_hemacias = df['hemacias'].mean()
mediana_hemoglobina = df['hemoglobina'].median()
desvio_padrao_hematocrito = df['hematocrito'].std()

print("Média de hemácias:", media_hemacias)
print("Mediana de hemoglobina:", mediana_hemoglobina)
print("Desvio padrão de hematócrito:", desvio_padrao_hematocrito)
```
